```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 1. Define the theoretical model
def sensor_model(h, A, B, C):
    return A / (B * h + C)

# 2. Your actual data
h = np.array([0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8])  # height in cm
Vout = np.array([942.0, 890.0, 828.8, 796.3, 707.4, 647.6, 607.6, 565.8])  # in mV

# 3. Curve fitting
initial_guess = [1000, 1, 1]
params, covariance = curve_fit(sensor_model, h, Vout, p0=initial_guess)
A_fit, B_fit, C_fit = params

# 4. Generate fitted curve
h_fit = np.linspace(min(h), max(h), 200)
V_fit = sensor_model(h_fit, A_fit, B_fit, C_fit)

# 5. Plotting
plt.figure(figsize=(8, 5))
plt.scatter(h, Vout, color='blue', label='Measured Data')
plt.plot(h_fit, V_fit, color='red', label=f'Fitted Curve: V = {A_fit:.2f} / ({B_fit:.2f}h + {C_fit:.2f})')
plt.xlabel('Height (cm)')
plt.ylabel('Output Voltage (mV)')
plt.title('Calibration Curve: Voltage vs Height')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Print parameters
print(f"Fitted Parameters:\n A = {A_fit:.4f}\n B = {B_fit:.4f}\n C = {C_fit:.4f}")
```

## Sensitivity



*   Mention that sensitivity decreases with height due to the inverse relationship. Theoretically at h = oo sensitivty goes to zero, but practically when after h>=2.5cm we cant see a much of variation in Vout

*   Peak sensitivity occurs at the smallest measurable distance (highest slope).



```python
# Define sensitivity function
def sensitivity(h, A, B, C):
    return - (A * B) / (B * h + C)**2

# Plotting Sensitivity vs Height
S_vals = sensitivity(h_fit, A_fit, B_fit, C_fit)

plt.figure(figsize=(8, 5))
plt.plot(h_fit, S_vals, color='green', label='Sensitivity dV/dh')
plt.xlabel('Height (cm)')
plt.ylabel('Sensitivity (mV/cm)')
plt.title('Sensitivity vs Height')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Measured data (example values  replace with yours)
h = np.array([0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8])
V = np.array([942.0, 890.0, 828.8, 796.3, 707.4, 647.6, 607.6, 565.8])  # Measured Vout

# Theoretical model
def model(h, A, B, C):
    return A / (B * h + C)

# Initial guess for A, B, C
initial_guess = [1000, 1, 1]

# Curve fitting
params, covariance = curve_fit(model, h, V, p0=initial_guess)
A_fit, B_fit, C_fit = params

# Compute fitted values
V_fit = model(h, A_fit, B_fit, C_fit)

# Residuals and R
residuals = V - V_fit
SS_res = np.sum(residuals**2)
SS_tot = np.sum((V - np.mean(V))**2)
R_squared = 1 - (SS_res / SS_tot)

# Plot measured vs fitted
plt.figure(figsize=(8, 5))
plt.scatter(h, V, label='Measured Data', color='blue')
plt.plot(h, V_fit, label=f'Fitted Model\n$V=\\frac{{{A_fit:.1f}}}{{{B_fit:.3f}h + {C_fit:.3f}}}$\n$R^2$ = {R_squared:.4f}', color='red')
plt.xlabel('Height (cm)')
plt.ylabel('Voltage (mV)')
plt.title('Linearity & Curve Fitting')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print results
print(f"Fitted Parameters:\nA = {A_fit:.3f}, B = {B_fit:.3f}, C = {C_fit:.3f}")
print(f"R (Goodness of Fit): {R_squared:.4f}")
```

## Hysteresis

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
heights = [0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8]
descending_voltage = [954.0, 902.2, 838.6, 808.2, 711.6, 655.0, 613.8, 579.2]
ascending_voltage = [942.0, 877.8, 819.0, 784.4, 703.2, 640.2, 601.4, 552.4]

# Create the figure and axis
plt.figure(figsize=(10, 6))

# Plot both curves
plt.plot(heights, descending_voltage, 'bo-', linewidth=2, label='Descending')
plt.plot(heights, ascending_voltage, 'ro-', linewidth=2, label='Ascending')

# Add arrows to show direction
for i in range(len(heights)-1):
    # Arrow for descending curve
    plt.annotate("", xy=(heights[i+1], descending_voltage[i+1]),
                xytext=(heights[i], descending_voltage[i]),
                arrowprops=dict(arrowstyle="->", color="blue", alpha=0.3))

    # Arrow for ascending curve
    plt.annotate("", xy=(heights[i+1], ascending_voltage[i+1]),
                xytext=(heights[i], ascending_voltage[i]),
                arrowprops=dict(arrowstyle="->", color="red", alpha=0.3))

# Connect the curves to form a closed loop
plt.plot([heights[-1], heights[-1]], [descending_voltage[-1], ascending_voltage[-1]], 'k--', alpha=0.5)
plt.plot([heights[0], heights[0]], [descending_voltage[0], ascending_voltage[0]], 'k--', alpha=0.5)

# Fill the hysteresis loop area
plt.fill_between(heights, descending_voltage, ascending_voltage, alpha=0.2, color='gray')

# Add labels and title
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Voltage (mV)', fontsize=12)
plt.title('Voltage Hysteresis Loop', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Calculate the hysteresis area
hysteresis_area = sum([(d - a) for d, a in zip(descending_voltage, ascending_voltage)])
plt.annotate(f'Hysteresis Area: {hysteresis_area:.1f} mVcm',
             xy=(0.5, 0.05),
             xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

# Show the plot
plt.tight_layout()
plt.show()

# Print the difference between descending and ascending values
print("Differences between Descending and Ascending Voltage Values:")
for h, d, a in zip(heights, descending_voltage, ascending_voltage):
    print(f"At height {h} cm: {d - a:.1f} mV")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Heights and voltage readings from Table 2
heights_cm = np.array([0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8])
V_descending = np.array([954.0, 902.2, 838.6, 808.2, 711.6, 655.0, 613.8, 579.2])
V_ascending = np.array([942.0, 877.8, 819.0, 784.4, 703.2, 640.2, 601.4, 552.4])

# Calculate absolute hysteresis
hysteresis_abs = np.abs(V_descending - V_ascending)

# Calculate full-scale span
V_max = max(np.max(V_descending), np.max(V_ascending))
V_min = min(np.min(V_descending), np.min(V_ascending))
full_scale_span = V_max - V_min

# Calculate percentage hysteresis
hysteresis_percent = hysteresis_abs / full_scale_span * 100

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(heights_cm, hysteresis_percent, marker='o', linestyle='-', color='purple')
plt.xlabel('Height (cm)')
plt.ylabel('Hysteresis Error (%)')
plt.title('Hysteresis Error vs Height')
plt.grid(True)
plt.tight_layout()
plt.show()

# Print numerical values
for h, hyst in zip(heights_cm, hysteresis_percent):
    print(f"Height = {h:.1f} cm => Hysteresis Error  {hyst:.2f} %")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example model parameters (replace with actual if needed)
A = 70840.3588
B = 42.3284
C = 54.9725

# Heights in cm
heights = np.array([0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8])

# Measured voltages in mV (user's example data)
V_measured = np.array([942,890,828.8,796.3,707.4,647.6,607.6,565.8])

# Theoretical voltages using the model
V_theoretical = A / (B * heights + C)

# Compute error metrics
abs_error = np.abs(V_measured - V_theoretical)
rel_error_percent = 100 * abs_error / V_theoretical
accuracy_percent = 100 - rel_error_percent

# Aggregate metrics
rmse = np.sqrt(np.mean((V_measured - V_theoretical)**2))
mape = np.mean(rel_error_percent)
mean_accuracy = np.mean(accuracy_percent)

# Plot measured vs theoretical
plt.figure(figsize=(10, 6))
plt.plot(heights, V_measured, 'o-', label='Measured Voltage')
plt.plot(heights, V_theoretical, 's--', label='Theoretical Voltage')
plt.title("Measured vs Theoretical Output Voltage")
plt.xlabel("Height (cm)")
plt.ylabel("Voltage (mV)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Plot accuracy per height
plt.figure(figsize=(10, 5))
plt.bar(heights, accuracy_percent, width=0.05, color='teal', alpha=0.7)
plt.axhline(mean_accuracy, color='r', linestyle='--', label=f'Mean Accuracy  {mean_accuracy:.2f}%')
plt.title("Accuracy per Height")
plt.xlabel("Height (cm)")
plt.ylabel("Accuracy (%)")
plt.grid(True)
plt.legend()
plt.tight_layout()

# Show both plots
plt.show()

print("rmse: ",rmse,"\nmape: ",mape,"\nmean accuracy: ", mean_accuracy)
```
